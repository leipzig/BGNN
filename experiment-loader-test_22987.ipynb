Plot information about the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

full_path = params["image_path"] + "/count_per_species.csv"
species_count_table = pd.read_csv(full_path).rename(columns={"fileName": "count"})
print(species_count_table)
species_count_table.plot.bar(x='scientificName', y='count', rot=90, figsize=(25, 10))
plt.savefig(params["image_path"] + '/count_per_species.pdf')
```

```python
full_path = params["image_path"] + "/count_per_family_genis.csv"
family_genis_count_table = pd.read_csv(full_path).rename(columns={"fileName": "count"}).sort_values(by=['count'])
print(family_genis_count_table)
family_genis_count_table.groupby('Family').agg('sum').reset_index().plot.bar(x='Family', y='count', rot=90, figsize=(15, 10))
plt.savefig(params["image_path"] + '/count_per_family.pdf')
family_genis_count_table.groupby('Genus').agg('sum').reset_index().plot.bar(x='Genus', y='count', rot=90, figsize=(15, 10))
plt.savefig(params["image_path"] + '/count_per_genis.pdf')
```

Get count matrix

```python
import numpy as np
import seaborn as sns; sns.set()

species_genis_matrix = pd.read_csv(params["image_path"] + '/cleaned_metadata.csv', sep='\t').groupby(['scientificName', 'Genus']).size().unstack(fill_value=0)

print(species_genis_matrix)
fig, ax = plt.subplots(figsize=(25,25))
sns.heatmap(species_genis_matrix, annot=True, fmt="d", ax=ax, cmap="YlGnBu", cbar=False)
plt.savefig(params["image_path"] + '/species_genis_matrix.pdf')
species_genis_matrix.to_csv(params["image_path"]+ '/species_genis_matrix.csv', sep='\t')
```
